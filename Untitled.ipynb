## Imports

```python
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Input, Flatten, Dense, Dropout, BatchNormalization, Conv2D, MaxPooling2D
from tensorflow.keras.datasets import fashion_mnist
```

## Load Data

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(x_train[i+10], cmap='gray')
    plt.title(y_train[i+10])
```

```python
model = keras.Sequential([
    Input(shape=(28, 28, 1)),
    Conv2D(32, (3, 3), padding='same', activation='relu'),
    MaxPooling2D((2, 2), strides=2),
    Conv2D(64, (3, 3), padding='same', activation='relu'),
    MaxPooling2D(2),
    Conv2D(128, (3, 3), padding='same', activation='relu'),
    MaxPooling2D(2),
    Flatten(),
    Dense(100, activation='relu'),
    Dense(10, activation='softmax')
])
```

```python
model.summary()
```

```python
x_train = x_train / 255
x_test = x_test / 255
 
y_train_cat = keras.utils.to_categorical(y_train, 10)
y_test_cat = keras.utils.to_categorical(y_test, 10)

x_train = np.expand_dims(x_train, axis=3)
x_test = np.expand_dims(x_test, axis=3)
```

```python
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
his = model.fit(x_train, y_train_cat, batch_size=50, epochs=20, validation_data=(x_test, y_test_cat))
```

```python
plt.plot(his.history['loss'], label='Train Loss')
plt.plot(his.history['val_loss'], label='Val Loss')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.show();
```
